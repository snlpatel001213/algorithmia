```python
# https://stanford.edu/~shervine/blog/pytorch-how-to-generate-data-parallel
import torch 
from torch import nn
from torch.utils import data
import os
```

```python
class Dataset(data.Dataset):
    def __init__(self, list_images, labels):
        self.list_images = list_images
        self.labels = labels
    def __len__(self):
        return len (self.list_images)
    def __getitem__(self, index):
        file_name = self.list_images[index]
        return file_name
```

```python
image_list = os.listdir("data/train/")
labels = []
for i in image_list:
    if i[:3] == 'dog':
        labels.append(1)
    else:
        labels.append(0)
```

```python
training_set = Dataset(image_list, labels)
```

```python
training_generator = data.DataLoader(training_set, batch_size=64, shuffle=False, num_workers=6)
```

```python
for fn in training_generator:
    print(fn)
```
